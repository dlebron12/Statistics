likelihood is a Bayesian topic

$$f(\theta|x) = \Pi_{i=1}^{n} f(x_{i}|\theta)$$ 

When we consider the pdf and calculate the joint probability of the data, we are considering theta as fixed and x as the variable, when we consider the **likelihood** function, we are considering x to be the observed sample point and theta to be varying over all possible parameter values. 

```python
# pmf and pdf equally used. 

from scipy.stats import binom

def update_binomial(pmf, data):
    """Update the pmf using the binomial distribution"""
    win, trials = data
    probs = pmf.qs #proportions of wins
    likelihood = binom.pmf(win, trials, probs)
    pmf*= likelihood
    pmf.normalize()
```

```python
from empiricaldist import Pmf
import numpy as np
hypos = np.linspace(0,1,101) #proportion of heads all with equal probability
uniform2 = Pmf(1, hypos, name='uniform2') 
data = 140, 250 #140 heads out of 250.

#the probability of landing heads being fed is equal to the possible proportion of heads listed in the prior
uniform2.qs
```

```python
update_binomial(uniform2, data)
```

## Exercises

In Major League Baseball, most baseball players have a batting average between .200 and .330, which means that their probability of getting hit is between 0.2 and 0.33.  Suppose a player appearing in their first game gets 3 hits out of 3 attempts.  What is the posterior distribution for their probability of getting hit.


```python
probs = np.linspace(0.1,0.4, 101)

# make uniform as each probability of getting hit is equally likely.
prior = Pmf(1,probs)


likelihood = {
    'Y': probs,
    'N': 1-probs
}

dataset = 'Y'*25 + 'N'*75 # is about .25 this is to represent batting averages

for data in dataset:
    prior*=likelihood[data]
prior.normalize()

prior.plot(label='prior')
plt.xlabel('Probability of getting a hit')
plt.ylabel('PMF')
```

```python
posterior = prior.copy()

for data in 'YYY':
    posterior *= likelihood[data]

posterior.normalize()
```

```python
prior.plot(label='prior')
posterior.plot(label='posterior ')
plt.xlabel('Probability of getting a hit')
plt.ylabel('PMF')
```
