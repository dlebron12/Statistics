```python
import matplotlib.pyplot as plt

f, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])

_ = ax.legend( scatter.legend_elements()[0], 
              iris.target_names, loc="lower right", 
              title="Classes")
```

## Mahalanobis distance

```python
X = iris['data']
X[:10]
```

```python
import numpy as np
from scipy.spatial.distance import mahalanobis

mean = np.mean(iris.data, axis = 0)
print(mean)
cov = np.cov(iris.data.T)

#Calculate mahalanobis distance
mahalanobis_dist = [mahalanobis(x, mean, np.linalg.inv(cov)) for x in iris.data]
```

```python
import matplotlib
from matplotlib.colors import ListedColormap

norm = plt.Normalize(min(mahalanobis_dist) , max(mahalanobis_dist))
plt.scatter(iris.data[:, 0], iris.data[:, 1], c = mahalanobis_dist,
           cmap = 'plasma')

# add a color bar
plt.colorbar()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.title('Mahalanobis Distance for Iris Dataset')
plt.show()

```
