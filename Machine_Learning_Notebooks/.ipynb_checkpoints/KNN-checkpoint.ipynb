Even though the points are multidimensional, the distance between them is still a scalar, or a single value.

```python
X = abalone.drop('Rings',axis = 1).values
Y = abalone['Rings'].values
```

```python
new_X = np.array([ 0.569552, 0.446407, 0.154437, 1.016849, 0.439051, 0.222526, 0.291208])

#distances between this new data point and others in the dataset
distances = np.linalg.norm(X - new_X, axis=1)
```

```python
# Find the data points with lowest distance to new X 
k = 3
nearest_neighbor_ids = distances.argsort()[:k]
nearest_neighbor_ids
```

```python
nearest_neighbor_rings = Y[nearest_neighbor_ids]
nearest_neighbor_rings.mean()
```

## KNN for Classification

you use the mode to see which label is the most frequent in the neighboring points; majority vote wins as classification for that new point.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, Y, 
                                                    test_size=0.2, 
                                                    random_state=12345)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor(n_neighbors=3)
knn_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
train_preds = knn_model.predict(X_train)
mse = mean_squared_error(y_train, train_preds)
rmse = sqrt(mse)
rmse
```

```python
test_preds = knn_model.predict(X_test)
mse = mean_squared_error(y_test, test_preds)
rmse = sqrt(mse)
rmse
```

the RMSE measures the average error of the predicted age, so you can interpret this as having, on average, an error of 1.65 years
